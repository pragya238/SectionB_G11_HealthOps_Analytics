```python

import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().resolve().name == 'notebooks' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from scripts.etl_pipeline import run_pipeline
```

```python
file_names = ['hospital_1.csv', 'hospital_2.csv', 'hospital_3.csv']

for fn in file_names:
    RAW_PATH = PROJECT_ROOT / f'data/raw/{fn}'
    PROCESSED_PATH = PROJECT_ROOT / f'data/processed/cleaned_{fn}'
    hospital_label = f"Hospital {fn.split('_')[1].split('.')[0]}"
    
    # Clean using ETL script run_pipeline
    clean_df = run_pipeline(RAW_PATH, PROCESSED_PATH, hospital_label)
    
    print(f'Processed {fn} -> {PROCESSED_PATH.name} | Cleaned shape: {clean_df.shape}')
```
